```python
### now
df.iloc[0].genres
```

```python
## I need this formate ['action', 'Adventure', 'Fantasy', 'Science fiction']

## input dic is a string, so let's convert it
import ast
ast.literal_eval  ## this function take string list and convert into original list

def convert(dic):
   L = []
   for i in ast.literal_eval(dic):
      L.append(i["name"])
      # print(i, end='')
   return L

# a = df.iloc[0].genres
# convert(a)

```

```python
df['genres'] = df['genres'].apply(convert)
df.head()
```

```python
## Need same formate for keywords variable 
df.keywords[0]
```

```python
## as there is also name key, so apply same function 
df['keywords'] = df.keywords.apply(convert)
```

```python
df.head(1)
```

```python
# df.overview[0]
# df.cast[0]
# df.crew[0]
```

```python
## so first of all, I will get at most 3 actors name from cast columns not more than that
## as there is also an "name" key, so i will use upper function


## input dic is a string, so let's convert it
import ast
ast.literal_eval  ## this function take string list and convert into original list

def fetch_Actors(dic):
   L = []
   count= 0
   for i in ast.literal_eval(dic):
      if count<3:
         L.append(i["name"])
         count+=1
      # print(i, end='')
      else:
         break

   return L

# a = df.iloc[0].genres
# convert(a)

```

```python
df.cast = df.cast.apply(fetch_Actors)
```

```python
## for crew I need just director name so let's see 
def take_dir(dic):
   L = []
   for i in ast.literal_eval(dic):
      if i['job'] == 'Director':
         L.append(i["name"])
         break

   return L
```

```python
df.crew =df.crew.apply(take_dir)  # bravo
```

```python
## now let's convert overview into list, I will need a list not string
df.overview = df.overview.apply(lambda x: x.split())
```

```python
## So far DataFrame is 
df.head()
```

```python
## as I want to have one paragraph of all these 4 column, for making tokens 
# here one problems is if i tokenize this, it will split each string into new small sting.
# like Same Worthington is one string, but it will be split into two string, which might confuse our model. I will remove this space
```

```python


# df['cast'] = df['cast'].apply(lambda x:[i.replace(" ", "") for i in x]) ## for single list
# df.head()

def rem_space(x):  ## for 4 columns
      L = []
      for i in x:
            L.append(i.replace(" ", ""))
      return L



a = df.iloc[:, 2:]  
names =a.columns  # getting names of columns

for i in names:
      df[i] = df[i].apply(rem_space)

df.head(2)
```

```python
## Now I will make a Tag columns where I will concatenate these columns for generating tags
df['tags'] = df.genres + df.keywords + df.overview + df.cast + df.crew
# df.tags[0]
```

```python
## let's drop rest which have been concatenated
df.drop(columns=['genres', 'keywords', 'cast', 'overview', 'crew'], inplace=True)
```

```python
# now let's convert tags into string for tokenization 
df.tags = df.tags.apply(lambda x: " ".join(x))
```

## Discussion
- as I need to compare each record tag with another record tag
- I will vectorized the text 
- why?
- As I want to recommend the similar movie, so by vectorization, The movie which will be closest to that movie it will be recommended
- it is called text-vectorizing, we will use bag of words for this. there are others like tfidf, wordvec etc

```python
# now we will apply stemming to get the dictionary form of words, as I was getting same words in get_feature_name() function of cv
import nltk
from nltk.stem.porter import PorterStemmer  ## stemming 
ps = PorterStemmer()
```

```python
# ps.stem("eating")  --> eat

def stem(text):
   L = []
   for i in text.split():  ## will take single values, so for this convert into list
      L.append(ps.stem(i))
   return " ".join(L)
      
df['tags'] = df.tags.apply(stem)
```

```python
#so for converting tags into vector
from sklearn.feature_extraction.text import CountVectorizer 
cv   = CountVectorizer(max_features=5000, stop_words='english') # stop_words will be removed 
# max_feature = finding frequent word upto 5000
```

```python
vector = cv.fit_transform(df['tags']).toarray()
vector[0]
# first movie form corpus (a large string)
```

```python
cv.get_feature_names() # these are the most frequent words in that tag
```

```python
## now each word is fresh form

# now for comparing each matrix with another, we will find the distance btw them with cosine similarity function of Sk-learn 
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
# it's actually measuring the distance of each record with all 4806 records
# that's way it has same rows and columns 
```

```python
similarity[2] # it's showing the similarity btw 2 and rest 
# this is the similarity matrix 
```

```python
## now task is, I will give movie, A function should return almost 5 similar movies name
# first I will find the index in my data
# second I will get it's similarity, and will sort in descending order
# we show top 5 similar movies
```

```python
index =df[df.title == 'Avatar'].index[0]  ## way of getting index
```

```python
sorted(similarity[index], reverse=True) # here problem is, if i sort, I will lose the position of the similar movie. so here we will use enumerate function, which gives us index with value of each record
```

```python
enumerate(similarity[index]) # --> it will give us object 
list(enumerate(similarity[index]))
sorted(list(enumerate(similarity[index])), reverse=True) # but this is compariing on the base of fist element, but we want with second element 

sorted(list(enumerate(similarity[index])), reverse=True, key= lambda x: x[1]) # here we go, now it is comparing with the second element which is 1

sorted(list(enumerate(similarity[index])), reverse=True, key= lambda x: x[1])[1:6] # as I just need top 5 records
```

```python
def recommend(movie):
   movie_index = df[df.title == movie].index[0]
   distance =sorted(list(enumerate(similarity[movie_index])), reverse=True, key= lambda x: x[1])[1:6]
   for i in distance:
      print(df.iloc[i[0]].title) 
      # print(i[0])
   
```

```python
print(recommend('Avatar'))

#so for getting the title of movie 
print(df.iloc[1216].title) 
```
